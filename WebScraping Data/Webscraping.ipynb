```python
import requests
from bs4 import BeautifulSoup
```

```python
class content_scraper:

    def __init__(self,url_l):
        self.url_l = url_l
        self.response = requests.get(self.url_l)
        self.soup = BeautifulSoup(response.content,"html.parser")

    def get_title(self):
        title = soup.find("title")
        return title.text
    
    def headers(self):
        headers = soup.find_all("h1",class_="firstHeading mw-first-heading")
        return headers[0].text
    
    def get_other_headers(self):
        other_headers = soup.find_all("div",class_="mw-heading mw-heading2")
        oth_headers = [header.text for header in other_headers]
        return oth_headers
    
    def get_para_content(self):
        p_content = []
        for p in ptags:
            p_content.append(p.text)
        paras = "\n".join(p_content)
        return paras
    
    def save_para(self):
        with open("paras_cotent","w",encoding="utf-8") as file:
            file.write(self.paras)
    
    def get_image_links(self):
        img_links = []
        for links in img_tags:
            link = links.get("href")
            img_links.append(self.url_l+link)
        return img_links
    
    def get_table(self):
        table_tags = soup.find_all("table",class_="wikitable")
        dfs = []
        for tab in table_tags:
            df = pd.read_html(str(tab))[0]
            display(df)
            dfs.append(df)
        return dfs
```
