### Inheritance


```python
class person:

    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def introduce(self):
        print(f"Hi, I am {self.name} and I am {self.age} years old")
```

```python
p1 = person('raman',45)
```

```python
p1.name
```

```python
p1.age
```

```python
p1.introduce()
```

```python
class employee(person):
    pass
```

```python
emp1 = employee('thaman',34)
```

```python
emp1.name
```

```python
emp1.age
```

```python
emp1.introduce()
```

```python
class employee(person):

    def __init__(self,company):
        self.company = company
    
    def work(self):
        print(f"Hi, I am working at {self.company}")
```

```python
emp2 = employee('wipro')
```

```python
emp2.company
```

```python
emp2.name
```

```python
class employee(person):

    def __init__(self,name,age,company):
        super().__init__(name,age)
        self.company = company
    
    def work(self):
        super().introduce()
        print(f"Hi, I am working at {self.company}")
```

```python
emp2 = employee('suman',38,'accenture')
```

```python
emp2.age
```

```python
emp2.name
```

```python
emp2.introduce()
```

```python
emp2.company
```

```python
emp2.work()
```
